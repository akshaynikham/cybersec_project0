```python
import pandas as pd
import matplotlib.pyplot as plt

raw_data = pd.read_csv('C:/Users/aksha/Desktop/CyberSecurityProjects/Data/raw_sensor_data.csv')
results = pd.read_csv('C:/Users/aksha/Desktop/CyberSecurityProjects/Data/trend_results.csv')

m = results['m'][0]
c = results['c'][0]

plt.scatter(raw_data['Time'], raw_data['Voltage'], color='gray', alpha=0.5, label='Raw Data(Noisy)')

x_line = raw_data['Time']
y_line = m * x_line + c

plt.plot(x_line, y_line, color='red', linewidth=2, label=f'Trend: y={m:.2f}x+{c:.2f}')

plt.legend()
plt.title('Least Squares: C Engine + Python Visuals')
plt.show()
```
